```python
import pandas as pd
import os

def load_statistics(model_dir):
    """
    Load statistics from CSV files in the model directory.

    Parameters:
    model_dir (str): Path to the directory containing model CSV files.

    Returns:
    list: A list of tuples where each tuple contains the model name and a pandas DataFrame with the statistics.
    """
    stats = []
    for file in os.listdir(model_dir):
        if file.endswith('_losses.csv'):
            model_name = file.replace('_losses.csv', '')
            file_path = os.path.join(model_dir, file)
            df = pd.read_csv(file_path)
            stats.append((model_name, df))
    return stats

# Example usage
model_dir = "networks/L1raw"
stats = load_statistics(model_dir)
```

```python
def analyze_statistics(stats):
    """
    Analyze the loaded statistics.

    Parameters:
    stats (list): A list of tuples where each tuple contains the model name and a pandas DataFrame with the statistics.
    """
    for model_name, df in stats:
        print(f"Statistics for model: {model_name}")
        print(df.describe())  # Print summary statistics
        print("\n")

# Example usage
analyze_statistics(stats)
```

```python
import matplotlib.pyplot as plt

def visualize_statistics(stats):
    """
    Visualize the loaded statistics.

    Parameters:
    stats (list): A list of tuples where each tuple contains the model name and a pandas DataFrame with the statistics.
    """
    for model_name, df in stats:
        plt.figure(figsize=(12, 6))
        plt.title(f"Total Loss Distributions for {model_name}")
        
        for dataset_type in df['dataset'].unique():
            subset = df[df['dataset'] == dataset_type]
            plt.hist(subset['total_loss'], bins=30, alpha=0.5, label=f"{dataset_type} Total Loss")
            plt.xlim(0.00, 0.2)
        
        plt.xlabel('Total Loss')
        plt.ylabel('Frequency')
        plt.legend(title='Dataset')
        plt.show()

# Example usage
visualize_statistics(stats)
```

```python
# Access and analyze elements
for model_name, df in stats[0:1]:
    print(f"Statistics for model: {model_name}")
    
    # Access the first row
    print("First row:\n", df.iloc[0])
    
    # Access a specific element
    print("Total loss of the first row:", df.loc[0, 'total_loss'])
    
    # Access a column
    total_loss_column = df['total_loss']
    print("Total loss column:\n", total_loss_column.head())
    
    # Sort the DataFrame by 'total_loss'
    sorted_df = df.sort_values(by='total_loss')
    print("Sorted DataFrame by total_loss:\n", sorted_df.head())
    
    # Filter rows where dataset is 'train'
    train_df = df[df['dataset'] == 'train']
    print("Training dataset:\n", train_df.head())
    
    # Filter rows where total_loss is greater than 0.5
    high_loss_df = df[df['total_loss'] > 0.5]
    print("High loss rows:\n", high_loss_df.head())

    # Visualize the total loss distribution
    plt.figure(figsize=(12, 6))
    plt.title(f"Total Loss Distribution for {model_name}")
    
    for dataset_type in df['dataset'].unique():
        subset = df[df['dataset'] == dataset_type]
        plt.hist(subset['total_loss'], bins=30, alpha=0.5, label=f"{dataset_type} Total Loss")
    
    plt.xlabel('Total Loss')
    plt.ylabel('Frequency')
    plt.legend(title='Dataset')
    plt.show()
```
